```python
def password_validation(password):
    lower = 0 
    upper = 0
    number = 0
    special = 0
    
    for i in password:
        if i.islower():
            lower += 1
            
        elif i.isupper():
            upper += 1
            
        elif i.isdigit():
            number += 1
            
        #elif i in "~!@#$%^&*()_+"
        elif not (i.isidentifier()):
            special += 1
            
    if lower >= 1 and upper >= 1 and number >= 1 and special >= 1 and len(password) >= 8:
        return True
    else:
        return False
    
    
def signup():
    user_name = input("Enter user name: ")
    password = getpass("Enter password: ")
    security_code = getpass("Enter security code: ")
    
    if password_validation(password):
        print("Initial amount of 2000 ₹ /- is required to activate your account")
        print("If you are agree enter yes(y) otherwise no(n): ")
        choice = input("Enter your choice y/n: ").lower()
        
        if choice == 'y':
            acc_no = list(data.keys())[-1] + 1
            balance = 2000
            data[acc_no] = [user_name, password, balance, security_code]
            print("Account created succefully")
            print("Your Account number is: ",acc_no)
            
        else:
            print("sorry we can not create account for you bacause initial amount is 2000 ₹ /-")
            print("Thanks for giving your time...")
            
    else:
        print("Password is not valid...")     
        
def login():
    acc_no = int(input("Enter your account number: "))
    password = getpass("Enter your password: ")
    
    if acc_no in data.keys() and password == data[acc_no][1]:
        print("login successfully...")
        
        while True:
            print("1. Credit")
            print("2. Debit")
            print("3. Change password")
            print("4. Check balance")
            print("5. Show Details")
            print("6. logout")
            
            choice = int(input("Enter your choice: "))
            
            if choice == 1:
                credit_amount = float(input("Enter amount for credit: "))
                data[acc_no][2] += credit_amount
                print(f"Amount created successfully..")
                
                
            elif choice == 2:
                debit_amount = float(input("Enter amount for debit: "))
                if (data[acc_no][2] - debit_amount >= 2000):
                    data[acc_no][2] -= debit_amount
                    print(f"Amount debited successfully...")
                    
                else:
                    print("Insufficient Balance...")
                    
            elif choice == 3:
                old_password = getpass("Enter old password: ")
                if old_password == data[acc_no][1]:
                    new_password = getpass("Enter new password: ")
                    if password_validation(new_password):
                        data[acc_no][1] = new_password
                        print("Password change successfully...")
                    
                    else:
                        print("password is not valid...")
                        
                else:
                    print("old password does not match.. Try again...")
                    
            elif choice == 4:
                password = getpass("Enter password: ")
                if password == data[acc_no][1]:
                    print(f"Your account balance is: {data[acc_no][2]} ₹")
                
                else:
                    print("Wrong password...")
                    print("Try again...")
                    
                    
            elif choice == 5:
                password = getpass("Enter password: ")
                if password == data[acc_no][1]:
                    print("Account Details: ")
                    print(f"Account number: {acc_no}")
                    print(f"User name: {data[acc_no][0]}")
                    
                else:
                    print("Wrong password...")
                    print("Try again...")
                    
            elif choice == 6:
                print("logout successfully...")
                break
                
    else:
        print("wrong account no / password...")
        
        
def forget_password():
    acc_no = int(input("Enter your account no: "))
    security_code = input("Enter your security code: ")
    
    if acc_no in data.keys() and security_code == data[acc_no][3]:
        new_password = getpass("Enter new password: ")
        if password_validation(new_password):
            data[acc_no][1] = new_password
            print("Password changed successfully...")
        else:
            print("Password is not valid...")
           
    else:
        print("Account no. or security code does not match")
            
while True:
    
    print("Welcome to The Bank Application".center(100, "*"))
    print("1 Login")
    print("2 Signup")
    print("3 Forget Password")
    print("4 Exit")
    
    query = int(input("Enter your query: "))
    
    if query == 1:
        login()
        
    elif query == 2:
        signup()
        
    elif query == 3:
        forget_password()
    
    elif query == 4:
        print("Successfully Exit")
        break
    
    else:
        print("Invalid input")         
```

```python
# Sample library management system using functions

# Initialize an empty dictionary to store books and their availability
library_books = {}

def add_book():
    book_title = input("Enter the title of the book: ")
    book_author = input("Enter the author of the book: ")
    if book_title in library_books:
        print("Book already exists in the library.")
    else:
        library_books[book_title] = book_author
        print(f"Book '{book_title}' by {book_author} added to the library.")

def remove_book():
    book_title = input("Enter the title of the book you want to remove: ")
    if book_title in library_books:
        del library_books[book_title]
        print(f"Book '{book_title}' removed from the library.")
    else:
        print("Book not found in the library.")

def display_books():
    if not library_books:
        print("The library is empty.")
    else:
        print("Books available in the library:")
        for title, author in library_books.items():
            print(f"{title} by {author}")

def library_system():
    while True:
        print("\nWelcome to the Library Management System")
        print("1. Add a book")
        print("2. Remove a book")
        print("3. Display all books")
        print("4. Exit")
        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            add_book()
        elif choice == '2':
            remove_book()
        elif choice == '3':
            display_books()
        elif choice == '4':
            print("Thank you for using the Library Management System.")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

# Run the library system
library_system()
```
